```python
#####################################
"""
  convert json to csv
"""
####################################
import json
import pandas as pd

file_path = "/content/json_data.json"

try:
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)
        print("JSON data loaded successfully.")

        rows = []
        for result in data.get("results", []):
            rows.append({
                "locationId": result.get("id"),
                "locality": result.get("locality")
            })

        df = pd.DataFrame(rows)
        print("\nDataFrame created:")
        print(df.head())

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except json.JSONDecodeError:
    print(f"Error: Could not decode JSON from the file '{file_path}'. Please check the file content.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
############################
"""
  SAVE TO CSV
"""
###########################



df.to_csv("locations.csv", index=False)
print("DataFrame successfully saved to locations.csv")
```

```python
import pandas as pd
import os

# Read the locations.csv file
try:
    locations_df = pd.read_csv("locations.csv")
    print("locations.csv loaded successfully.")
except FileNotFoundError:
    print("Error: locations.csv not found. Please make sure you have created it.")
    exit() # Exit if the file is not found

# Define the year
year = 2025

# Define the base destination directory
base_destination_dir = "openaq_data"
os.makedirs(base_destination_dir, exist_ok=True)


# Iterate through each locationId and execute the aws s3 cp command
for index, row in locations_df.iterrows():
    location_id = row["locationId"]

    # Define the specific directory for this locationId
    location_destination_dir = os.path.join(base_destination_dir, str(location_id))
    os.makedirs(location_destination_dir, exist_ok=True)

    s3_command = f"aws s3 cp --no-sign-request --recursive s3://openaq-data-archive/records/csv.gz/locationid={location_id}/year={year}/ {location_destination_dir}"

    print(f"Executing command for locationId {location_id}: {s3_command}")
    # Execute the command in the shell
    os.system(s3_command)

print("Data download process completed for all locations.")
```
